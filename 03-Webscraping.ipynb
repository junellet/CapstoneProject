# Combining Code

```python
#Create video list
video_urls= []

# have to reload the page at least 112 times to generate over 1000 video URLs (112 reloads x 9 videos per reload)
num_reloads = 112

for _ in range(num_reloads):
    # Save the URL of the webpage we want to scrape to a variable
    website = 'https://randommer.io/random-youtube-videos'

    # Send a get request and assign the response to a variable
    video_results = requests.get(website)

    # Turn the undecoded content into a Beautiful Soup object and assign it to a variable
    video_soup = BeautifulSoup(video_results.content, 'html.parser')

    # Find all the <a> tags with href attributes containing the YouTube link
    videos = video_soup.find_all('a', href=True)

    # Loop through each video in our list of video tags
    for video in videos:
        # Try the following
        try:
            # Check if 'href' contains the string youtube etc
            if 'www.youtube.com/watch?v=' in video['href']:
                #print(video['href'])
                # Append the YouTube link to the list
                video_urls.append(video['href'])

        # If an error is returned from the above then we continue with the loop
        except:  # It's good practice to specify the exception type, e.g., `except KeyError:`
            continue
```
